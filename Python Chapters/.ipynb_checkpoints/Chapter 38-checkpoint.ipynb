# Chapter 38 How to get information out of a list within a dictionary

```python
details = {
  "nickname": "Bajwa",
  "married": "no",
  "careers": ["Web Developer", "Software Engineer", "CEO"],
}
if "Web Developer" in details["careers"]:
  print("yes")
else:
  print("no")
```

```python
details = {
  "nickname": "Bajwa",
  "married": "yes",
  "careers": ["Web Developer", "Web Designer", "Singer", "Rider"],
}
if details["married"] == "yes" and "Web Developer" in details["careers"]:
    print("Uh-Yes!")
else:
    print("no")

```
